```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Data
student=np.array(["Tim","John","Amit","Jacob","Karl","Gary"])
marks=np.array([85,59,99,77,67,95])
```

```python
# Plot a bar chart
plt.bar(student, marks)

# the labels for x and y co-ordinates
plt.xlabel("Student")
plt.ylabel("Marks")

# Plot title
plt.title("Students marks")
```
