```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Data
student=np.array(["Tim","John","Amit","Jacob","Karl","Gary"])
marks=np.array([85,59,99,77,67,95])
```

```python
# the labels for x and y co-ordinates
plt.xlabel("Student",fontsize=15)
plt.ylabel("Marks",fontsize=15)

# Plot title
plt.title("Students marks",fontsize=20)

# Customized color for different student
c=['lightblue', 'blue', 'purple', 'red', 'green','orange']

# Plot a bar chart

# plt.bar(student, marks,width=0.5,color=c,align="edge",edgecolor="black",linewidth=3,linestyle=":",alpha=0.5)
plt.bar(student, marks,width=0.5,color=c)

plt.legend("marks")
plt.show()
# align->edge or center
```

```python
# Data
x=["python","c","c++","Java"]
y=[60,80,90,40]
z=[30,40,60,20]

width=0.2
# to show parllel
p=np.arange(len(x))
p1=[j+width for j in p]

# the labels for x and y co-ordinates
plt.xlabel("Language",fontsize=15)
plt.ylabel("Number",fontsize=15)

# Plot title
plt.title("Platform",fontsize=20)

# Plot a bar chart
plt.bar(p, y,width,color="green",label="popularity")
plt.bar(p1, z,width,color="blue",label="popularity")

plt.xticks(p+width/2,x,rotation=30)
plt.legend()
plt.show()
```
