```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# df=pd.read_csv(r"E:\Learn Python\Data Science\Learn\Data\iris_update.csv")
df=pd.read_csv("../../../Data/iris_update.csv")
df.head(5)
```

```python
sns.pairplot(data=df)
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
wcss=[]

for i in range(2,20):
    km=KMeans(n_clusters=i,init="k-means++")    
    km.fit(df)
    wcss.append(km.inertia_)  ### value of  wcss
```

```python
plt.figure(figsize=(10,5))
plt.plot([i for i in range(2,20)],wcss,marker="o")
plt.xlabel("Number of Clusters")
plt.xticks([i for i in range(2,20)])
plt.ylabel("WCSS")
# plt.grid(axis="x")
plt.grid()
plt.show()
```

### Hence from 3 cluster cluster is downfall so, 3 number will be choose for clustering

```python
kmn=KMeans(n_clusters=3)
df["predict"]=kmn.fit_predict(df)
df
```

```python
sns.pairplot(data=df,hue="predict",palette="Accent")
# plt.savefig("predict_iris.jpg")
plt.show()
```
