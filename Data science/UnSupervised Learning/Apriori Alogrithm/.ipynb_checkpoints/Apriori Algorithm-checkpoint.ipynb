```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("../../../Data/groceries-Apriori.csv")
df.head(5)
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
market=[]
for i in range(0,df.shape[0]):
    cus=[]
    for j in df.columns:
         if type(df[j][i])==str:
             cus.append(df[j][i])
    market.append(cus)
```

```python
market
```

```python
l=[]
for i in market:
    for j in i:
        l.append(j)
```

```python
l
```

```python
import collections
```

```python
p=collections.Counter(l)
p
```

```python
p.keys()
```

```python
p.values()
```

```python
d={"Item Name":p.keys(),"values":p.values()}
```

```python
pd.DataFrame(d).sort_values(by=["values"],ascending=False)
```

```python
from mlxtend.preprocessing.transactionencoder import TransactionEncoder
```

```python
tr=TransactionEncoder()
tr.fit(market)
```

```python
df1=pd.DataFrame(tr.transform(market),columns=tr.columns_)
df1
```

```python
from mlxtend.frequent_patterns import apriori
```

```python
apriori(df1,min_support=0.07,use_colnames=True,max_len=3).sort_values(by=["support"])
```
