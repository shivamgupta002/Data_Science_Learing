# Ordinal Encoding (using Sckit leran and using Map function)

```python
odf=pd.DataFrame({"Size":["s","m","l","xl","s","m","l","s","s","l","xl","m"]})
odf.head(4)
```

```python
ord_data=[["s","m","l","xl"]]
```

### Using Sckit learn

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe=OrdinalEncoder(categories=ord_data)
oe.fit(odf[["Size"]])
```

```python
odf["Size_en"]=oe.transform(odf[["Size"]])
```

```python
odf
```
