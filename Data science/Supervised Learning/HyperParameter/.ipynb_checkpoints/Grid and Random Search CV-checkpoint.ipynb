```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("../../../Data/placement.csv")
df.head(6)
```

```python
x=df.iloc[:,:-1]
y=df["package"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dt=DecisionTreeRegressor()
dt.fit(x_train,y_train)
```

```python
dt.score(x_train,y_train)*100,dt.score(x_test,y_test)*100,
```

```python
from sklearn.model_selection import GridSearchCV,RandomizedSearchCV
```

```python
hp={ "criterion":['squared_error',"friedman_mse", "absolute_error","poisson"],
    "splitter":['best',"random"],
    "max_depth":[i for i in range(2,20)]
   }
```

## Grid Search CV

```python
gd=GridSearchCV(DecisionTreeRegressor(),param_grid=hp)
gd.fit(x_train,y_train)
```

```python
gd.best_params_
```

```python
gd.best_score_*100
```

```python
dt1=DecisionTreeRegressor(criterion="absolute_error",max_depth=4,splitter="random")
dt1.fit(x_train,y_train)
```

```python
dt1.score(x_train,y_train)*100,dt1.score(x_test,y_test)*100,
```

## Randomized Search CV

```python
rd=RandomizedSearchCV(DecisionTreeRegressor(),param_distributions=hp,n_iter=20)
rd.fit(x_train,y_train)
```

```python
rd.best_params_
```

```python
rd.best_score_*100
```

```python
dt2=DecisionTreeRegressor(criterion="poisson",max_depth=4,splitter="random")
dt2.fit(x_train,y_train)
```

```python
dt2.score(x_train,y_train)*100,dt2.score(x_test,y_test)*100,
```

## Result

```python
print("Normal Decision :",dt.score(x_train,y_train)*100,dt.score(x_test,y_test)*100)
print("After Grid Search CV :",dt1.score(x_train,y_train)*100,dt1.score(x_test,y_test)*100)
print("After Randomized Search CV Decision :",dt2.score(x_train,y_train)*100,dt2.score(x_test,y_test)*100)
```
