```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
```

```python
df=pd.read_csv("../../../../data/placement_update.csv")
df.head(5)
```

```python
df.isnull().sum()
```

### For check normal distribution

```python
plt.figure(figsize=(6,3))
sns.kdeplot(data=df["cgpa"])
plt.show()
```

```python
plt.figure(figsize=(6,3))
sns.kdeplot(data=df["score"])
plt.show()
```

```python
plt.figure(figsize=(6,3))
sns.scatterplot(x="cgpa",y="score",data=df,hue="placed")
plt.show()
```

```python
x=df.iloc[:,:-1]
y=df["placed"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.2)
```

```python
from sklearn.naive_bayes import GaussianNB,MultinomialNB,BernoulliNB
```

### Gaussian

```python
gnb=GaussianNB()
gnb.fit(x_train,y_train)
```

```python
gnb.score(x_test,y_test)*100,gnb.score(x_train,y_train)*100
```

```python
plt.figure(figsize=(6,3))
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=gnb)
plt.show()
```

### Multinomial

```python
mnb=MultinomialNB()
mnb.fit(x_train,y_train)
```

```python
mnb.score(x_test,y_test)*100,mnb.score(x_train,y_train)*100
```

```python
plt.figure(figsize=(6,3))
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=mnb)
plt.show()
```

```python
gnb.predict([[7.19,26]])
```

### Bernoulli

```python
bnb=BernoulliNB()
bnb.fit(x_train,y_train)
```

```python
bnb.score(x_test,y_test)*100,bnb.score(x_train,y_train)*100
```

```python
plt.figure(figsize=(6,3))
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=bnb)
plt.show()
```
