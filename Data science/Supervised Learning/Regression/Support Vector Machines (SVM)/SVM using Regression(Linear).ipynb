## For Linear Data

```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("../../../../data/placement.csv")
df.head(5)
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(6,3))
sns.scatterplot(x="cgpa",y="package",data=df)
plt.show()
```

```python
x=df[["cgpa"]] ## machine learning algorith work on 2-d array
y=df[["package"]]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.2)
```

```python
from sklearn.svm import SVR ## SVR is Support Vector Regression
```

```python
### kernel= "linear","rbf","poly","sigmoid","precomputed" 

sv=SVR() ## By default rbf used
sv.fit(x_train,y_train)
```

```python
sv1=SVR(kernel="linear")
sv1.fit(x_train,y_train)
```

```python
sv2=SVR(kernel="poly") 
sv2.fit(x_train,y_train)
```

```python
sv3=SVR(kernel="sigmoid") 
sv3.fit(x_train,y_train)
```

```python
print("Testing Score")

print("rbf",sv.score(x_test,y_test)*100)
print("linear",sv1.score(x_test,y_test)*100)
print("poly",sv2.score(x_test,y_test)*100)
print("sigmoid",sv3.score(x_test,y_test)*100)
```

```python
print("Training Score")

print("rbf",sv.score(x_train,y_train)*100)
print("linear",sv1.score(x_train,y_train)*100)
print("poly",sv2.score(x_train,y_train)*100)
print("sigmoid",sv3.score(x_train,y_train)*100)
```

### Here Linear and then rbf,poly kernel is good

```python
### Linear Graph 
sns.scatterplot(x="cgpa",y="package",data=df)
plt.plot(df["cgpa"],sv1.predict(x),color="red")
plt.show()
```
