### For Linear Model

```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
```

```python
df=pd.read_csv("../../../../data/CollegePlacement.csv")
df.head(5)
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(7,3))
sns.scatterplot(x="cgpa",y="score",data=df,hue="placed")
plt.show()
```

```python
x=df.iloc[:,:-1]
y=df["placed"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=15,test_size=0.2)
```

```python
from sklearn.svm import SVC ## SVC-> Support vector classifier
```

```python
sv=SVC(kernel="linear")
sv.fit(x_train,y_train)
```

```python
sv.score(x_test,y_test)*100
```

```python
sv.score(x_train,y_train)*100
```

Here the differnce between training and testing datset will be more so there will be some error.
we can change number of random state

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=sv)
plt.show()
```

### We can chanage kernel also to get better answer

```python
### kernel= "linear","rbf","poly","sigmoid","precomputed" 
sv=SVC(kernel="linear")
sv.fit(x_train,y_train)
```

```python
sv1=SVC(kernel="poly") 
sv1.fit(x_train,y_train)
```

```python
sv2=SVC(kernel="sigmoid") 
sv2.fit(x_train,y_train)
```

```python
# sv3=SVC(kernel="precomputed") 
sv3=SVC(kernel="rbf") ## rbf= radial basic function
sv3.fit(x_train,y_train)
```

```python
print("Testing Score")
print("linear",sv.score(x_test,y_test)*100)
print("poly",sv1.score(x_test,y_test)*100)
print("sigmoid",sv2.score(x_test,y_test)*100)
print("precomputed",sv3.score(x_test,y_test)*100)
```

```python
print("Training Score")
print("linear",sv.score(x_train,y_train)*100)
print("poly",sv1.score(x_train,y_train)*100)
print("sigmoid",sv2.score(x_train,y_train)*100)
print("precomputed",sv3.score(x_train,y_train)*100)
```

# Hence the best kernel will be "Poly" and then "Linear"

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=sv1)
plt.show()
```
