```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("../../../data/Social_Network_Ads.csv")
df.head(10)
```

# Imbalance Data Set (Check)

```python
df["Purchased"].value_counts()
```

### In this data model is biased

```python
x=df.iloc[:,:-1]
y=df["Purchased"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
lr.predict([[19,19000]])
```

```python
lr.predict([[45,26000]]) 
```

### It shows wrong pedictions hence our model is biased

# Imblearn

## UnderSampling

```python
from imblearn.under_sampling import RandomUnderSampler
```

```python
ru=RandomUnderSampler()
ru_x,ru_y=ru.fit_resample(x,y)
```

```python
ru_x
```

```python
ru_y
```

```python
ru_y.value_counts()
```

### In this data model is not biased (Because we use random under sampling)

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(ru_x,ru_y,random_state=42,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
lr.predict([[19,19000]])
```

```python
lr.predict([[48,29000]]) 
```

## Over Sampling

```python
from imblearn.over_sampling import RandomOverSampler
```

```python
ro=RandomOverSampler()
ro_x,ro_y=ro.fit_resample(x,y)
```

```python
ro_y.value_counts()
```

### In this data model is not biased (Because we use random Over sampling)

```python
from sklearn.model_selection import train_test_split
x_train1,x_test1,y_train1,y_test1=train_test_split(ro_x,ro_y,random_state=42,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train1,y_train1)
```

```python
lr.score(x_test1,y_test1)*100
```

```python
lr.predict([[19,19000]])
```

```python
lr.predict([[48,29000]]) 
```
