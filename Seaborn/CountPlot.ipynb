# Count Plot

## Differnce between countPlot and bar plot

```python
# Countplot plots the count of the number of records by category
# Barplot plots a value or metric for each category (bydefault, barplot plots the mean of a variable, by category)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
var=sns.load_dataset("tips")
var
```

```python
sns.barplot(x="sex",y="size",data=var)
plt.show()
```

```python
sns.countplot(x="sex",data=var)
plt.show()
```

```python
sns.countplot(x="sex",data=var,hue="smoker",palette="Accent")
plt.show()
```

```python
sns.countplot(x="sex",data=var,hue="smoker",color="blue")
plt.show()
```

```python
sns.countplot(x="sex",data=var,hue="smoker",saturation=0.3)
plt.show()
```
