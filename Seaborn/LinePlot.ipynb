## Line Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
var=[1,2,3,4,5,6,7]
var_1=[2,3,4,1,6,7,8]
x_1=pd.DataFrame({"var":var,"var_1":var_1})
```

```python
plt.plot(var,var_1)
plt.show()
```

```python
# sns.lineplot(x="var",y="var_1",data=x_1,) #show label when use double quotes
sns.lineplot(x=var,y=var_1,data=x_1,)
plt.show()
```

```python
# y_1=sns.load_dataset("penguins")
y_1=sns.load_dataset("penguins").head(20)
# y_1=sns.load_dataset("penguins").tail(20)
y_1
```

```python
# sns.lineplot(x="bill_length_mm",y="bill_depth_mm",data=y_1)
# sns.lineplot(x="bill_length_mm",y="bill_depth_mm",data=y_1,hue="sex",size=30)
# sns.lineplot(x="bill_length_mm",y="bill_depth_mm",data=y_1,hue="sex",style="sex")
sns.lineplot(x="bill_length_mm",y="bill_depth_mm",data=y_1,hue="sex",style="sex",
             palette="Accent",markers=["o","^"],dashes=False,legend="auto") #legend will be of True or False(legend=False)
plt.grid()
plt.title("Penguine")
plt.show()
```
