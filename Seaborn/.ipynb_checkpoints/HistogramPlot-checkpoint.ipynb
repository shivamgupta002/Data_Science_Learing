## Histogram Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
var=sns.load_dataset("penguins")
var
```

```python
sns.displot(var["flipper_length_mm"])
# sns.displot(var["sex"])
plt.show()
```

```python
sns.displot(var["flipper_length_mm"],bins=[170,180,190,200,210,220,230,240])
plt.show()
```

```python
sns.displot(var["flipper_length_mm"],bins=[170,180,190,200,210,220,230,240],kde=True) #kde=kernel density estimator
plt.show()
```

```python
sns.displot(var["flipper_length_mm"],bins=[170,180,190,200,210,220,230,240],rug=True,color="green")
plt.show()
```

```python
sns.displot(var["flipper_length_mm"],rug=True,color="red",log_scale=True)
plt.show()
```

### var["flipper_length_mm"]: Plots the distribution of the "flipper_length_mm" column.
### bins=[170,180,190,200,210,220,230,240]: Specifies custom bin intervals for the histogram.
### kde=True: Adds a smooth KDE (Kernel Density Estimate) curve to show the distribution's shape.
### rug=True: Adds small marks along the x-axis for each data point, indicating where values fall.
### color="green": Sets the plot color to green.
### log_scale=True: Uses a logarithmic scale on the x-axis.
