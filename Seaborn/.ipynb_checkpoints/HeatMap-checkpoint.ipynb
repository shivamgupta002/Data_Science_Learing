## HeatMap

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
data=sns.load_dataset("anagrams")
data
```

```python
x=data.drop(columns=["attnr"],axis=1).head(5)
x
```

```python
sns.heatmap(x)
```

```python
var=np.linspace(1,10,20).reshape(4,5)
var
```

```python
sns.heatmap(var)
plt.show()
```

```python
sns.heatmap(var,vmin=0,vmax=18)
plt.show()
```

```python
sns.heatmap(var,vmin=0,vmax=12,cmap='Blues',annot=True) #annot represents that value will be present or not
plt.show()
```

```python
var1=np.linspace(1,10,10).reshape(2,5)
var1
```

```python
sns.heatmap(var1,vmin=0,vmax=12,cmap='Blues',annot=True) 
plt.show()
```

```python
ar=np.array([["a0","a1","a2","a3","a4"],["b0","b1","b2","b3","b4"]])
ar
```

```python
sns.heatmap(var1,vmin=0,vmax=12,cmap='Blues',annot=ar,fmt="s") #fmt is like type of data 
plt.show()
```

```python
# y={"fontsize":15,"color":"g"}
y={'fontsize':10, 
   'fontstyle':'italic',  
   'color':"k",
   'alpha':0.6, 
   'rotation':"vertical",
   'verticalalignment':'center',
   'backgroundcolor':'w'
  }

sns.heatmap(var1,vmin=0,vmax=12,cmap='Blues',annot=ar,fmt="s",annot_kws=y,linewidth=10) #line width is width of line graph(particular part) 
plt.show()
```

```python
y={"fontsize":15,"color":"g"}
sns.heatmap(var1,vmin=0,vmax=12,cmap='Blues',annot=ar,fmt="s",annot_kws=y,linewidth=10,linecolor="y") 
plt.show()
```

```python
sns.heatmap(var1,vmin=0,vmax=12,cmap='Blues',annot=ar,fmt="s",cbar=False) 
plt.show()
```

```python
sns.heatmap(var1,vmin=0,vmax=12,cmap='Blues',annot=ar,fmt="s",xticklabels=False,yticklabels=False) 
#xticklabels,yticklabels shows that want labels in x and y axis respectively 
plt.show()
```

```python
v=sns.heatmap(var1,vmin=0,vmax=12,cmap='Blues',annot=ar,fmt="s",xticklabels=False,yticklabels=False)
v.set(xlabel="python",ylabel="coding")
sns.set(font_scale=1.5) #0.0-9.0 
plt.show()
```
