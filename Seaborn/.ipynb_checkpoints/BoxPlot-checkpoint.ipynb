# Box Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
var=sns.load_dataset("tips")
var
```

```python
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=var)
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(x="total_bill",y="day",data=var)
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=var,hue="sex")
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=var,hue="sex",palette="Accent")
plt.show()
```

```python
sns.set(style="whitegrid")
o=["Fri","Sun","Thur","Sat"]
sns.boxplot(x="day",y="total_bill",data=var,hue="sex",order=o)
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=var,hue="sex",showmeans=True)
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=var,showmeans=True,meanprops=({"marker":"+","markeredgecolor":"red"}))
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=var,hue="sex",showmeans=True,linewidth=3)
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(x="total_bill",y="day",data=var,orient="h") # for orient both value should be y axis will be numerical and x axis will be numerical 
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(data=var,orient="h") # for orient both value should be y axis will be numerical and x axis will be numerical 
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(x=var["total_bill"])
# sns.boxplot(y=var["total_bill"])
plt.show()
```
