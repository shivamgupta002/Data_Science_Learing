## Scatter Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
var=sns.load_dataset("penguins")
var
```

```python
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=var)
plt.show()
```

```python
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=var,hue="sex")
plt.show()
```

```python
var_20=sns.load_dataset("penguins").head(20)
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=var_20,hue="sex",style="sex")
plt.show()
```

```python
var_20=sns.load_dataset("penguins").head(20)
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=var_20,hue="sex",style="sex",palette='Accent',alpha=0.6)
plt.show()
```

```python
m={"Male":"<","Female":"o"}
var_20=sns.load_dataset("penguins").head(20)
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=var_20,hue="sex",style="sex",palette='Accent',markers=m)
plt.show()

# hue: Differentiates points by color based on the "sex" column.
# style: Differentiates points by marker shape based on the "sex" column.
# palette: Specifies the color scheme ('Accent' palette used here).
# markers: A dictionary mapping categories ("Male" to <, "Female" to o).
```

```python
var_20 = sns.load_dataset("penguins").head(20)

# Scatter plot
sns.scatterplot(
    x="bill_length_mm",
    y="bill_depth_mm",
    data=var_20,
    hue="sex",
    style="sex",
    palette="Accent",
    legend="brief"  # Use 'brief' or 'full' to display legend
)
# Customize the legend
plt.legend(
    title="Gender",              # Legend title
    labels=["Female", "Male"],   # Custom labels
    # loc="upper left",            # Legend position
    loc="lower right",            # Legend position
    fontsize=10                  # Font size
)

plt.show()
```
