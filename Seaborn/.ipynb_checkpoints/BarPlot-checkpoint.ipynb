## Bar Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
val=sns.load_dataset("penguins")
val
```

```python
sns.barplot(x="island",y="bill_depth_mm",data=var)
# sns.barplot(x=var.island,y=var.bill_depth_mm)
plt.show()
```

```python
sns.barplot(x="island",y="bill_depth_mm",data=var,hue="sex")
plt.show()
```

```python
order_1=["Dream","Torgersen","Biscoe"]
sns.barplot(x="island",y="bill_depth_mm",data=var,hue="sex",order=order_1)
plt.show()
```

```python
order_1=["Dream","Torgersen","Biscoe"]
sex_order=["Female","Male"]
sns.barplot(x="island",y="bill_depth_mm",data=var,hue="sex",order=order_1,hue_order=sex_order)
plt.show()
```

```python
order_1=["Dream","Torgersen","Biscoe"]
sex_order=["Female","Male"]
sns.barplot(x="island",y="bill_depth_mm",data=var,hue="sex",order=order_1,hue_order=sex_order,errorbar=('ci', 0)) 
#ci is black line on garph which is from 0-100
plt.show()
```

```python
order_1=["Dream","Torgersen","Biscoe"]
sex_order=["Female","Male"]
sns.barplot(x="island",y="bill_depth_mm",data=var,hue="sex",order=order_1,hue_order=sex_order,errorbar=('ci', 30),n_boot=3) 
#ci is black line on garph which is from 0-100
plt.show()

# x="island": The column from var to be plotted on the x-axis (categorical variable, "island").
# y="bill_depth_mm": The column from var to be plotted on the y-axis (numerical variable, "bill_depth_mm").
# data=var: The DataFrame (var) containing the data for the plot.
# hue="sex": Adds color differentiation for the bar groups based on the "sex" column (e.g., Female vs. Male).
# order=order_1: Customizes the order of x-axis categories to ["Dream", "Torgersen", "Biscoe"].
# hue_order=sex_order: Customizes the order of hues (grouped bars) to ["Female", "Male"].
# errorbar=('ci', 30): Displays confidence intervals (CIs) with a width of 30%. You can adjust the percentage.
# n_boot=3: Sets the number of bootstrap iterations for error estimation to 3 (default is 1000). Lower values make calculations faster but less precise.
```

```python
sns.barplot(x="body_mass_g",y="bill_depth_mm",data=var,orient="h") # orient paramter work only when x and y both are numeric type
plt.show()
```

```python
color_1 = ["blue", "green", "red"]
sns.barplot(x="island", y="bill_depth_mm", data=var, palette=color_1) 
plt.show()
```

```python
color_1 = ["green", "red"]
sns.barplot(x="island", y="bill_depth_mm", data=var,hue="sex", palette=color_1,saturation=0.5,err_kws={'color': 'br','linewidth':'3'})  
plt.show()

# palette=color_1: Specifies the color palette for the hue groups.
# saturation=0.5: Adjusts the intensity of the bar colors. A lower value like 0.5 makes the colors appear lighter.
# err_kws={'color': 'br', 'linewidth':'3'}: Customizes the error bars:
# color='br': This is invalid as br is not a color; you should use valid color codes like "blue", "black", or "#FF5733".
# linewidth='3': Sets the thickness of the error bar lines to 3.
```

```python
color_1 = ["green", "red"]
sns.barplot(x="island", y="bill_depth_mm", data=var,hue="sex", palette=color_1,saturation=0.5,capsize=0.2) # capsize is used for cap on the graph
plt.show()
```

```python
color_1 = ["blue", "orange"]
sns.barplot(x="island", y="bill_depth_mm", data=var,hue="sex", palette=color_1,dodge=False)  #dodge help to prevent overlape
plt.show()
```

```python
color_1 = ["blue", "orange"]
sns.barplot(x="island", y="bill_depth_mm", data=var,hue="sex", palette=color_1,alpha=0.6)  #alpha use for blur the graph 0-1
plt.show()
```

```python
color_1 = ["blue", "orange"]
# sns.set(style="dark")
sns.set(style="darkgrid")
sns.barplot(x="island", y="bill_depth_mm", data=var,hue="sex", palette=color_1,alpha=0.6)  
plt.show()
```
