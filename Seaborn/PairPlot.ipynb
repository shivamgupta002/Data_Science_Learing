## Pair Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# var=sns.load_dataset("tips")
var=sns.load_dataset("tips").head(10)
var
```

```python
sns.pairplot(var)
plt.show()
```

```python
sns.pairplot(var,hue="sex")
plt.show()
```

```python
sns.pairplot(var,vars=["total_bill","tip"],hue="sex",hue_order=["Female","Male"])
plt.show()
```

```python
sns.pairplot(var,vars=["total_bill","tip"],hue="sex",palette="Accent")
plt.show()
```

```python
# sns.pairplot(var,hue="sex",x_vars=["total_bill","tip"])
sns.pairplot(var,hue="sex",y_vars=["total_bill","tip"])
plt.show()
```

```python
sns.pairplot(var,vars=["total_bill","tip"],kind="kde") #kind-> 'scatter','kde','hist','reg'
plt.show()
```

```python
sns.pairplot(var,vars=["total_bill","tip"],kind="kde",diag_kind="hist") #diag_kind-> 'scatter','kde','hist','reg'
plt.show()
```

```python
sns.pairplot(var,vars=["total_bill","tip"],hue="sex",markers=["*","<"])
plt.show()
```
