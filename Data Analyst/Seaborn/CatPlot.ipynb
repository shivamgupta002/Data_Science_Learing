```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
var=sns.load_dataset("tips")
var
```

```python
var20=sns.load_dataset("tips").head(20)
var20
```

```python
sns.catplot(x ='size', y ='tip', data = var)
plt.show()
```

```python
sns.catplot(x ='size', y ='tip', data = var20,hue="sex")
plt.show()
```

```python
sns.catplot(x ='size', y ='tip', data = var20,hue="sex",palette="Accent",height=3)
plt.show()
```

```python
sns.catplot(x ='size', y ='tip', data = var20,hue="sex",kind="strip")
plt.show()
```

```python
sns.catplot(x ='size', y ='tip', data = var20,hue="sex",kind="swarm")
plt.show()
```

```python
sns.catplot(x ='size', y ='tip', data = var20,hue="sex",kind="point")
plt.show()
```

```python
sns.catplot(x ='size', y ='tip', data = var20,hue="sex",kind="bar")
plt.show()
```

```python
sns.catplot(x ='size', data = var20,hue="sex", kind="count")
plt.show()
```

```python
sns.catplot(x ='size', y ='tip', data = var20, hue="sex", kind="box")
plt.show()
```

```python
sns.catplot(x ='size', y ='tip', data = var20, hue="sex", kind="violin")
plt.show()
```

```python
sns.catplot(x ='size', y ='tip', data = var20, hue="sex", kind="boxen")
plt.show()
```
