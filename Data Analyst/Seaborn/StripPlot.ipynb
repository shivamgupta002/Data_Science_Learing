# Strip Plot

### It is basiclly a scatter plot that differentiates different categories. Strip plots are considered a good alternative to a box plot or a violin plot. 

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
var=sns.load_dataset("tips")
var
```

```python
sns.stripplot(x="day",y="total_bill",data=var)
plt.show()
```

```python
sns.stripplot(x="day",y="total_bill",data=var,hue="sex")
plt.show()
```

```python
sns.stripplot(x="day",y="total_bill",data=var,hue="sex",palette="hsv") #palette->color
plt.show()
```

```python
sns.stripplot(x="day",y="total_bill",data=var,hue="sex",linewidth=2)
plt.show()
```

```python
sns.stripplot(x="day",y="total_bill",data=var,hue="sex",linewidth=2,edgecolor="red")
plt.show()
```

```python
sns.stripplot(x="day",y="total_bill",data=var,hue="sex",jitter=1) #jitter ->spread 
plt.show()
```

```python
sns.stripplot(x="day",y="total_bill",data=var,hue="sex",size=10)
plt.show()
```

```python
sns.stripplot(x="day",y="total_bill",data=var,hue="sex",marker=">")
plt.show()
```

```python
sns.stripplot(x="day",y="total_bill",data=var,hue="sex",alpha=0.5) #alpha lights the graph
plt.show()
```
