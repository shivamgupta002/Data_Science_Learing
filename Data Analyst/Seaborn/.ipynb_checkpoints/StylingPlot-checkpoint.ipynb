# Styling Plot in Seaborn

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
var=sns.load_dataset("tips")
var
```

```python
sns.boxplot(x=var["total_bill"])
plt.show()
```

```python
sns.barplot(x="day",y="total_bill",data=var)
plt.show()
```

```python
sns.barplot(x="day",y="total_bill",data=var)
plt.grid()
plt.show()
```

```python
# Seaborn Figure style
sns.set_style("whitegrid") #style must be one of white, dark, whitegrid, darkgrid, ticks
sns.barplot(x="day",y="total_bill",data=var)
plt.show()
```

```python
# Removing Axes Spines
sns.set_style("whitegrid")
sns.barplot(x="day",y="total_bill",data=var)
sns.despine()
plt.show()
```

```python
# Scale and Context
plt.figure(figsize=(12,3))
sns.barplot(x="day",y="total_bill",data=var)
plt.show()
```

```python
# Scale and Context
sns.set_context("poster",font_scale=1.5) #paper, notebook, talk, poster
sns.barplot(x="day",y="total_bill",data=var)
plt.show()
```

```python
# Scale and Context
sns.set_context("notebook",font_scale=0.9)
sns.barplot(x="day",y="total_bill",hue="sex",data=var,palette="cool")
plt.show()
```
