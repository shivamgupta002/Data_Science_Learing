## Violine Plot

### A violin plot plays a similar role as abox and whisker plot.It shows the distribution of quantitative data across several levels of one (or more) categorical varibales such that those distribution can be compared.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
var=sns.load_dataset("tips")
var
```

```python
sns.violinplot(x="day",y="total_bill",data=var)
plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",data=var,hue="time")
plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",data=var,linewidth=3)
plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",data=var,hue="sex",palette="Accent")
plt.show()
```

```python
sns.violinplot(x="time",y="total_bill",data=var)
plt.show()
```

```python
sns.violinplot(x="time",y="total_bill",data=var,order=["Dinner","Lunch"])
plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",data=var,hue="sex",palette="Accent",saturation=0.5)
plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",data=var,hue="sex",split=True)
plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",data=var,hue="sex",split=True,density_norm='count') # desnity_norms(scale)="count","area","count",width"
plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",data=var,hue="sex",inner="quart")
plt.show()
```

```python
sns.violinplot(y=var["sex"])
plt.show()
```
