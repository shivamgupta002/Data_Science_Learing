# Multiple Plots (FacetGrid)

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
var=sns.load_dataset("tips")
var
```

```python
fg=sns.FacetGrid(var)
fg.map(plt.scatter,"total_bill","tip")
plt.show()
```

```python
fg=sns.FacetGrid(var,col="sex")
fg.map(plt.hist,"total_bill")
# fg.map(plt.plot,"total_bill")
plt.show()
```

```python
fg=sns.FacetGrid(var,col="sex")
fg.map(plt.scatter,"total_bill","tip")
plt.show()
```

```python
fg=sns.FacetGrid(var,col="sex",hue="day")
fg.map(plt.scatter,"total_bill","tip").add_legend()
plt.show()
```

```python
fg=sns.FacetGrid(var,col="day",hue="sex")
fg.map(plt.scatter,"total_bill","tip").add_legend()
plt.show()
```

```python
fg=sns.FacetGrid(var,col="day",hue="sex")
fg.map(plt.bar,"total_bill","tip").add_legend()
plt.show()
```

```python
fg=sns.FacetGrid(var,col="day",hue="sex",palette="summer")
fg.map(plt.scatter,"total_bill","tip",edgecolor="red").add_legend()
plt.show()
```
