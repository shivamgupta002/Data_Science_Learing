```python
import matplotlib.pyplot as plt
import numpy as np 
```

```python
# Data
x=[1,2,3,4,5,6]
y=[2,2,5,4,3,6]
plt.stem(x,y)
# Line Style
# "_" for solid line
# "__" for Dashed line
#"_." for Dashed-dot line
# "." for Dotted line

plt.show()
```

```python
plt.stem(x,y,linefmt=".",markerfmt="ro",bottom=1,basefmt="green",orientation="horizontal")
```

```python
plt.stem(x,y,linefmt=".",markerfmt="ro",bottom=1,basefmt="green",label="python",orientation="vertical")
plt.legend()
```
