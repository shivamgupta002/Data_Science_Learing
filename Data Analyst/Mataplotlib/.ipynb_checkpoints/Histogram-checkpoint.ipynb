```python
# A histogram is a chart that shows how numeric data is distributed by grouping it into bins and displaying the bins as segmented columns.
# Bins: Each bar in a histogram represents a range of numeric values, called a bin or class
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Data
arr=np.array([10,13,14,45,65,119,23,24,24,35,119,76,98,66,88,99,43,54,76,32,133])
```

```python
# plot a histogram
# the bins paramter sets the bins i.e an integer or sequence
plt.hist(arr,bins=[0,20,40,60,80,100,120])

# Labels
plt.xlabel("Marks")
plt.ylabel("Student")
```
