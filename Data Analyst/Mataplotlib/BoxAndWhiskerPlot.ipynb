```python
import matplotlib.pyplot as plt
```

```python
# Data
x=[10,20,30,40,50,60,70]
```

```python
plt.boxplot(x)
plt.show()
```

```python
plt.boxplot(x,patch_artist=True)
plt.show()
```

```python
plt.boxplot(x,notch=True)
plt.show()
```

```python
plt.boxplot(x,vert=False)
plt.show()
```

```python
# Default width is approx equal to 0.2
plt.boxplot(x,widths=0.3)
plt.show()
```

```python
# while using labels it always write in list
plt.boxplot(x,labels=["python"],showmeans=True)
# plt.label()
plt.show()
```

```python
# Dot represnts the outlar
y=[10,20,30,40,50,60,70,120]
# plt.boxplot(y)

# to join the outlar use whis
# plt.boxplot(y,whis=3)

# sym change outlar color 
plt.boxplot(y,sym="g+",boxprops=dict(color="red"),capprops=dict(color="blue"),whiskerprops=dict(color="purple"),flierprops=dict(markeredgecolor="green"))

plt.show()
```

```python
# Dot represnts the outlar
x1=[20,40,50,30,70]

z=[x,x1,y]
plt.boxplot(z,labels=["puthon","c++","java"])

plt.show()
```
