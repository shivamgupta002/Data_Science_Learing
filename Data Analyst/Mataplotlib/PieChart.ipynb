```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Data
cricketer=np.array(["Tim","John","Amit","Jacob","Karl","Gary"])
runs=np.array([85,59,99,77,67,95])
```

```python
# Plot a pie chart
plt.pie(runs, labels=cricketer, autopct='%1.3f%%')

# Plot title
plt.title("Runs scored in a cricket match")
```
