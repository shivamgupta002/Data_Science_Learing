```python
# Data
data={
    "Cricket_Bat":["SG","BDM","SS","GM","Kookaburra","Spartan"],
    "MRP":[2000,2200,2600,2500,2700,2850],
    "Weight_Grams":[1100,1200,1250,1330,1480,1000]
}
```

```python
#DataFrame
dataFrame=pd.DataFrame(data)
```

```python
# Plot a line using the pyplot.plt() method

# The x and y co-ordinate are the columns of dataframe
plt.plot(dataFrame["Cricket_Bat"],dataFrame["MRP"])

# Add lables for x and y coordinate
plt.xlabel("Bat Company")
plt.ylabel("Price(USD)")

# Set the title on the left
plt.title("Bat Price depends on the Company",loc='center')
```

```python
# Add Legends in a graph
a=np.arange(5)
b=[2,4,6,8,10]
c=[5,6,7,8,9]
```

```python
# Create Plots
fig=plt.figure()
ax=plt.subplot()
ax.plot(a,b,'k--',label='Frequency')
ax.plot(a,c,'k:',label='Periods')

# Create a legend using teh metaplotlib Axes.legend() in python
ax.legend()
# ax.legend(loc="lower right")
# legend=ax.legend(loc="upper center",fontsize="x-small")

#font-size will be -> xx-small,x-small,small,medium,large,x-large,xx-large

# Set the background color of the legend 
# legend.get_frame().set_facecolor('blue')

# Plot Title
# plt.title("Frequency of a Signal")
plt.title("Frequency of a Signal",loc="center")
```
